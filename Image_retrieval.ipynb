```python
import pandas as pd
import numpy as np
import urllib
from PIL import Image
```

```python
# connecting to drive
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
airbnb_london_filtered = pd.read_csv("/content/gdrive/My Drive/Thesis/London_Data/airbnb_london_filtered_advanced.csv")

airbnb_london_filtered.head()
```

Saving all images in a seperate Folder in Drive with a name referencing their ID and creating a DF which indicates if the image could be retrieved (as this mechanism does not work for listings which are not online anymore).

```python

img_available = list(np.zeros(airbnb_london_filtered.shape[0])) ## is the image available

image_df = pd.DataFrame({'id': airbnb_london_filtered.id, 'img_available': img_available})
```

```python

for i in range(airbnb_london_filtered.shape[0]):

  if i % 20 == 0:
    print(f'Currently at {i/airbnb_london_filtered.shape[0] * 100} percent.')

  try: ## will not work if listing is down!

    url_to_be_retrieved = airbnb_london_filtered.picture_url[i]
    id_here = airbnb_london_filtered.id[i]

    urllib.request.urlretrieve(url_to_be_retrieved, "image_intermediate_save.jpg")
    img_intermediate = Image.open("image_intermediate_save.jpg")
    ## saving in lower quality for computational reasons
    img_intermediate.save(f"/content/gdrive/My Drive/Thesis/Image_data/img_{id_here}.jpg", quality = 10)

    # filling corresponding lists
    image_df.img_available[i] = 1

  except:
    pass

  # saving intermediate results:
  if i % 200 == 0:
    image_df.to_csv("/content/gdrive/My Drive/Thesis/London_Data/test_img_extra.csv", index = False)

image_df.to_csv("/content/gdrive/My Drive/Thesis/London_Data/test_img_extra.csv", index = False)
```
